# 7.1  데이터 수집 1 – 한 페이지 크롤링

### 7.1.2 다나와 검색 페이지 접속

```python
from selenium import webdriver
driver = webdriver.Chrome('chromedriver.exe')
url = "http://search.danawa.com/dsearch.php?query=무선청소기&tab=main"
driver.get(url)
```

### 7.1.3 다나와 검색 웹 페이지에서 상품 정보 가져오기
- 검색결과의 1페이지를 구성하는 HTML 정보를 가지고와서 BeautifulSoup으로 Read 

```python
from bs4 import BeautifulSoup

html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
```

- 페이지 소스 확인을 통해 상품정보는 태그명이 li이고 class값은 prod_item인 태그에 위치
- select함수를 통해 태그의 개수 확인

```python
prod_items_all = soup.select('li.prod_item')
len(prod_items_all)
```

- 페이지에서 찾고자하는 상품의 개수는 40개이지만, 현재 44개로 나옴
- 40개의 제품만 조회하도록 태그정보 수정

```python
prod_items = soup.select('div.prod_info')
len(prod_items)
```

```python
title = prod_items[0].select('div.prod_info > p.prod_name > a')
print(str(title))
```

- 상품명 정보만 가지고 오기 (1개의 상품이 더 조회되는데 어떤 상품인지 확인해보자)

```python
for i in range(len(prod_items)):
    title = prod_items[i].select('div.prod_info > p.prod_name > a')[0].text.strip()
    print(title)
```

- 상품의 spec 정보를 가지고오기

```python
spec_list = prod_items[0].select('div.spec_list')
print(spec_list)
```

```python
spec_list = prod_items[0].select('div.spec_list')[0].text.replace('\t','').replace('\n','')  
print(spec_list)
```

- 상품의 가격 정보를 가지고오기 (상품의 가격 중 1위 가격만 가지고 오기)

```python
price = prod_items_all[0].select('li.rank_one > p.price_sect > a > strong')[0].text.strip().replace(",", "")
price
```

- 반복문으로 검색 결과의 1페이지에 대한 상품 정보 추출

```python
for i in range(len(prod_items)):
    try:
        title = prod_items[i].select('div.prod_info > p.prod_name > a')[0]
        if not "productPot.ajax" in str(title):
            print(title.text.strip())
            spec_list = prod_items[i].select('div.spec_list')[0].text.replace('\t','').replace('\n','')  
            print(spec_list)
            price = prod_items_all[i].select('li.rank_one > p.price_sect > a > strong')[0].text.strip().replace(",", "")
            print(price +'\n')
    except:
        pass
```

- 상품 정보 태그에서 원하는 정보를 추출하는 함수

```python
def get_prod_items(prod_items):
    prod_data = []
    for i in range(len(prod_items)):
        try:
            title = prod_items[i].select('div.prod_info > p.prod_name > a')[0]
            if not "productPot.ajax" in str(title):
                title = title.text.strip()
                
                spec_list = prod_items[i].select('div.spec_list')[0].text.replace('\t','').replace('\n','')  
                
                price = prod_items_all[i].select('li.rank_one > p.price_sect > a > strong')[0].text.strip().replace(",", "")
                prod_data.append([title, spec_list, price])
        except:
            pass  
    return prod_data
```

```python
prod_items = soup.select('div.main_prodlist > ul.product_list > li.prod_item')
prod_data = get_prod_items(prod_items)
print(len(prod_data))
prod_data
```

# 7.2  데이터 수집 2 - 여러 페이지에 걸친 다나와 검색 페이지 크롤링

### 7.2.1 다나와 검색 결과 페이지 URL 분석
- 다나와 첫페이지 URL 및 2페이지 URL 분석

- 1페이지 진입시 : https://search.danawa.com/dsearch.php?k1=%EC%A7%84%EA%B3%B5%EC%B2%AD%EC%86%8C%EA%B8%B0&module=goods&act=dispMain
- 2페이지 진입시 : https://search.danawa.com/dsearch.php?query=%EC%A7%84%EA%B3%B5%EC%B2%AD%EC%86%8C%EA%B8%B0&originalQuery=%EC%A7%84%EA%B3%B5%EC%B2%AD%EC%86%8C%EA%B8%B0&volumeType=allvs&page=2&limit=40&sort=saveDESC&list=list&boost=true&addDelivery=N&recommendedSort=Y&defaultUICategoryCode=10325109&defaultPhysicsCategoryCode=72%7C80%7C81%7C0&defaultVmTab=4273&defaultVaTab=832666&tab=goods

```python
def get_search_page_url(keyword, page):
    return 'http://search.danawa.com/dsearch.php?query={}&volumeType=allvs&page={}&limit=40&sort=saveDESC&list=list&boost=true&addDelivery=N&tab=goods&tab=goods'.format(keyword, page)

keyword = '무선청소기'
page = 2
url = get_search_page_url(keyword, page)
print(url)
```

### 7.2.3 여러 페이지에 걸친 상품 정보 수집

```python
from selenium import webdriver
import time
from bs4 import BeautifulSoup
from tqdm.notebook import tqdm 


keyword = '무선청소기'
total_page = 10
prod_data_total = []
 
for page in tqdm(range(1, total_page + 1)):
    
    url = get_search_page_url(keyword, page)
    driver.get(url)
    
    time.sleep(5)
    
    
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    
    
    prod_items = soup.select('div#productListArea > div.main_prodlist > ul.product_list > li.prod_item')
    prod_item_list = get_prod_items(prod_items)
    
    
    prod_data_total = prod_data_total + prod_item_list
```

### 7.2.4 수집 데이터 저장 

```python
import pandas as pd
data = pd.DataFrame(prod_data_total)
data.columns = ['상품명', '스펙 목록', '가격']
data.to_excel('1_danawa_crawling_result.xlsx', index = False)
```
