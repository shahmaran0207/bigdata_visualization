## 크롬드라이버 활용하기

```python
from selenium import webdriver
driver=webdriver.Chrome('chromedriver.exe')
```

## 웹 페이지 접속

```python
url='https://search.naver.com/search.naver?where=news&sm=tab_jum&query=%EB%8D%B0%EC%9D%B4%ED%84%B0%EC%8B%9C%EA%B0%81%ED%99%94'
driver.get(url)
```

### 웹 페이지(HTML) 다운로드

```python
html=driver.page_source
html
```

```python
from bs4 import BeautifulSoup
soup=BeautifulSoup(html, 'html.parser')
soup.select('div.news_area > a')
```
