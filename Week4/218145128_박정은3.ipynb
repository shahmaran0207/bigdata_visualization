- Fare의 경우 사람들마다 값의 차이가 큰데, 어떤 식으로 구성된 것일까? 

1) 탑승 요금이 500이 넘는 승객은 1등급 클래스

```python
titanic[titanic['Fare']>500]
```

2) 탑승 요금이 0인 승객 중 20살 이상 승객 존재

```python
df=titanic[titanic["Fare"]==0]
df[df['Age']>=20]
```

3) 탑승 요금이 있는 승객 중 5살 이하 승객 존재

```python
df=titanic[titanic['Fare']>0]
df[df['Age']<=5]
```

- 그러면 왜 요금이 0인 승객이 있을까? 

```python
titanic[titanic['Fare']==0]
```

```python
titanic[titanic['Ticket']=='LINE']
```
