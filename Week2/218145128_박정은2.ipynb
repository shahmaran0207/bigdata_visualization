### NumPy

#### NumPy 사용 이유

- Python List vs. NumPy

 1) Python 리스트로 총점 구하기

```python
mid_scores=[10, 20, 30]
final_scores=[70,80,90]

total=[]

for i in range(len(mid_scores)):
    total.append(mid_scores[i]+final_scores[i])
total
```

 2) NumPy를 활용하여 총점 구하기

```python
import numpy as np

mid_scores=np.array([10, 20, 30])
final_scores=np.array([70, 80, 90])

total=mid_scores+final_scores
print('시험성적 합계', total)
print('시험성적 평균', total/2)
```

### Exercise (논리적인 인덱싱)

(실습) 여러 사람의 BMI를 계산해보고, 이중 25인 사람만 추출해보자


```python
import numpy as np

heights=[1.83, 1.76, 1.69, 1.86, 1.77, 1.73]
weights=[86,74,59,95,80,68]

np_heights=np.array(heights)
np_weights=np.array(weights)

bmi=np_weights/(np_heights**2)

print('대상자들의 키', np_heights)
print('대상자들의 몸무게', np_weights)
print('대상자들의 bmi', bmi)
print('bmi가 25넘는 사람의 bmi', bmi[bmi>25])
```

#### 다차원 배열

- 1차원 배열

```python
import numpy as np #numpy import
x=[1,2,3,4,5,6]

arr1=np.array(x, int) #array() 함수의 입력 값 리스트 X
                         #np.array(x)와 동일

print('생성된 numpy 배열 확인', arr1, type(arr1)) #생성된 numpy배열 확인

print('생성된 numpy 배열 자료형 확인', arr1.dtype) #생성된 numpy 배열 자료형

print('배열의 형태: shape={}, 배열의 차원: ndim={}, 배열의 원소개수: size={}'.format(arr1.shape, arr1.ndim, arr1.size))
```

### Exercise (2차원 배열 슬라이싱)

(실습) 2차원 배열에서 특정조건을 만족하는 행만 추출하기

```python
import numpy as np

players=[[170,76.4],
        [183, 86.2],
        [181, 78.5],
        [176, 80.1]]

np_players=np.array(players)

print('몸무게가 80이상인 선수 정보')
print(np_players[:,1]) #몸무게만 산출
print(np_players[:,1]>=80.0) #Boolean 형태의 Array추출
print(np_players[np_players[:,1]>=80.0])
print('키가 180이상인 선수 정보')
print(np_players[:,0]>=180.0) #Boolean형태의 Array추출
print(np_players[np_players[:,0]>=180.0])
```
