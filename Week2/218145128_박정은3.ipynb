### matplotlib 라이브러리

- matplotlib 라이브러리는 파이썬에서 2D 형태의 그래프, 이미지 등을 그릴 때 사용
- 실제 과학 컴퓨팅 연구 분야나 인공지능 연구 분야에서도 많이 활용
- 파이썬으로 데이터를 시각화하는 데는 matplotlib 라이브러리를 가장 많이 사용

- matplotlib 라이브러리 안에는 다양한 모듈이 존재, 그 중 pyplot모듈을 주로 사용
- matplotlib.pyplot는 너무 길고 복잡하기에 import할 때 plt라는 별명 사용

#### 1차원 데이터 생성 및 시각화

- Pyhton package 가져오기 및 matplotlib 출력 옵션 설정
  * %matplotlib inline : ipython의 “magic function” 중 하나로, matplotlib의 시각화 결과를 Ipython notebook 안에서 출력하는 함수

```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

- 임의의(Random) 값 40개로 이루어진 배열 변수 생성
  * np.random.standard_normal(생성숫자개수) : 평균 0 표준편차 1인 (가우시안) 표준정규분포 (np.random.randn과 유사)
  * np.random.standard_normal vs. np.random.randn: np.random.standard_normal은 turple을 입력으로 받을 수 있음
  * e.g. np.random.standard_normal((10, 2)): 10*2 matrix의 평균 0 표준편차 1인 표준정규분포의 값   
  * np.random.rand(생성숫자개수): 0~1사이의 균일분포

```python
import numpy as np

value=np.random.standard_normal(40)
value
```

- 선 Plot 그리기 및 x/y축 범위 설정
  * pyplot.plot() : 선이나 마커를 플롯하는 함수
  * pyplot.xlim() : x축의 범위를 설정하는 함수
  * pyplot.ylim() : y축의 범위를 설정하는 함수
  * numpy.min() : 값들 중 최소값을 구하는 함수
  * numpy.max() : 값들 중 최대값을 구하는 함수

```python
index=range(len(value)) #0부터 value 변수 데ㅣ터 길이 -1까지의 연속적인 값(0~39)
plt.plot(index, value)

plt.xlim(0, 20)
plt.ylim(np.min(value)-1, np.max(value)+1)
```

- Plot 옵션 설정 및 꾸미기
  * pyplot.figure() : plot의 새로운 figure(모양)을 생성하는 함수 
  * numpy.cumsum() : 누적 합계를 구하는 함수
  * pyplot.xlabel() : x축의 이름을 설정하는 함수
  * pyplot.ylabel() : y축의 이름을 설정하는 함수
  * pyplot.title() : 플롯의 제목을 설정하는 함수

```python
plt.figure(figsize=(7, 4))
plt.plot(value.cumsum(), 'b', linewidth=1.5)
plt.plot(value.cumsum(), 'ro')
plt.xlabel('index')
plt.ylabel('value')
plt.title('Line Plot 1')
```

#### 2차원 데이터 생성 및 시각화

- 임의의(Random) 값 이루어진 2차원 배열 변수 생성하기

```python
import numpy as np

value=np.random.standard_normal((30, 2))
value
```

- 2차원 배열 변수를 하나의 그래프에 그리기
  * 하나의 플롯 그래프 안에 두개의 데이터셋에 대한 그래프를 표현
  * 각각의 데이터셋을 라벨을 붙여서 구분
  * pyplot.grid(): grid(격자) 표시 유무 설정
  * pyplot.legend(): 범례 표시

```python
plt.figure(figsize=(10, 4))
plt.plot(value[:, 0], linewidth=1.5, label='1st')
plt.plot(value[:, 1], linewidth=1.5, label='2nd')
plt.plot(value, 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('index')
plt.ylabel('value')
plt.title('Line Plot 2')
```

#### SubPlot (다중그래프)

- 2차원 배열 변수를 이용한 다중그래프 (subplot) 그리기
  * 2개의 subplot 그래프 그리기
  * subplot() 함수를 이용하여 2개의 plot을 배열 형태로 분리
  * pyplot.subplot(nrows ncols plot_number):nrows = 전체 서브플롯 행의 개수,ncols = 전체 서브플롯 열 개수,plot_num: 특정 플롯의 번호
  * subplot(211): 2*1개의 서브플롯으로 이루어진 플롯의 첫번째 서브플롯 

```python
#1번쨰 
plt.figure(figsize=(10, 5))
plt.subplot(211)
plt.plot(value[:, 0], linewidth=1.5, label='1st')
plt.plot(value[:, 0], 'ro')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('Line Plot 3')

#2번째
plt.subplot(212)
plt.plot(value[:, 1], 'g', linewidth=1.5, label='2nd')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight') #axis(option): 축의 option 설정, 'tight': 모든 데이터를 볼 수 있도록 범위를 충분히 넓게 설정
plt.ylabel('value')
```

- 2차원 배열 변수를 이용한 다중그래프 (subplot) 그리기
  * 6개의 subplot 그래프 그리기

```python
#1번째
plt.figure(figsize=(13, 5))
plt.subplot(231)
plt.plot(value[:, 0], linewidth=1.5, label='1st')
plt.plot(value[:, 0], 'co')
plt.grid(True)
plt.legend(loc=0)
plt.ylabel('value')
plt.title('Line Plot 3')

#2번째
plt.subplot(232)
plt.plot(value[:, 0], 'g-', linewidth=1.5, label='1st')
plt.plot(value[:, 0], 'bD')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')

#3번째
plt.subplot(233)
plt.plot(value[:, 0], 'g', linewidth=1.5, label='1st')
plt.plot(value[:, 0], 'bD')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')

#4번째
plt.subplot(234)
plt.plot(value[:, 1], '*', linewidth=1.5, label='2nd')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')

#5번째
plt.subplot(235)
plt.plot(value[:, 1], 'b', linewidth=1.5, label='2nd')
plt.plot(value[:, 1], 'ms')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')

#6번째
plt.subplot(236)
plt.plot(value[:, 1], 'r--', linewidth=1.5, label='2nd')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.ylabel('value')
```

#### 막대그래프(bar(), barh())
- 입력 데이터의 값을 막대 모양의 길이로 나타낸 그래프 

```python
import numpy as np
import matplotlib.pyplot as plt

x=np.random.randint(1, 10, size=10)
y=np.random.randint(5, 20, size=10)

plt.title('Bar Graph')
plt.xlabel('x')
plt.ylabel('y')

plt.bar(x, y)
plt.show()
```

```python
plt.title('Bar Graph')
plt.xlabel('y')
plt.ylabel('x')

plt.barh(x, y,
        facecolor='red', #색
        edgecolor='black',
        alpha=0.3, #투명도
        linewidth=3) #bar 굵기
plt.show()
```

#### 파이차트(pie())
- 각 데이터가 차지하는 비율을 원으로 표현
  * 일변량 자료를 표현
  * 원에 각 범주에 해당되는 비율만큼 각도를 분할하여 표시 (해당 범주 각도 = 비율 *360°) 

```python
import matplotlib.pyplot as plt

plt.title('Pie Chart')
ratio=[35, 30,25,20,20]

plt.pie(ratio)
plt.show()
```

```python
ratio=[35,30,25,20,20]
labels=['Apple', 'Samsung', 'Google', 'Naver', 'Kakao']
plt.title('Pie Chart')
plt.pie(ratio, 
       labels=labels, #레이블 출력
       startangle=90,  # 시작 위치(반시계방향으로 그려짐)
       autopct='%.2f%%') #비율값 표시

plt.show()
```

#### 히스토그램(hist())
- 입력 값(전체 집단)의 분포 상태를 파악하거나 비교할 때 사용
  * bins: 연속적인 변수를 히스토그램에 나타내기 위해 데이터를 나누는 개수

```python
import matplotlib.pyplot as plt
import numpy as np

value=np.random.standard_normal((500,2)) #500*2개의 표준정규분포 따르는 Random 값
plt.figure(figsize=(12,7))
plt.hist(value, label=['1st', '2nd'], bins=30, color=('yellow', 'green'))
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram')
```

### Pandas에서 matplotlib사용

#### 1) Series에서 matplotlib 사용

```python
from matplotlib import pyplot as plt
import pandas as pd

month=['1', '2', '3', '4','5', '6']
block_chain=[10,30,20,40,70,60]

plt.plot(month, block_chain) #plot 생성
plt.savefig('My_block_chain_chain1.png')# 이미지로 저장
plt.show()
```

```python
block_chain={'1':10, '2':30, '3':20, '4':40, '5':70, '6':60}

bc=pd.Series(block_chain) #pd시리즈 생성

plt.plot(bc) #plot생성

plt.xlabel('month') #x축 이름
plt.ylabel('percent') #y축 이름
plt.title('My_Block_chain') #그래프 제목

plt.savefig('My_block_chain2.png') #이미지로 저장
plt.show()
```

#### 2) DataFrame에서 matplotlib 사용

```python
my_score=[[60,90,95],[80,75,100]]
subject=['kor', 'math', 'eng']
year=['2019', '2020']

df=pd.DataFrame(my_score, columns= subject, index=year)
df
```

```python
df.plot()
df.plot(subplots=True)
```

```python
df.plot(kind='bar')
df.plot.bar(subplots=True)
```

```python
df.plot(kind='pie', subplots=True )
df.plot(kind='pie', y='kor')
```
