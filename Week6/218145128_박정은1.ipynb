## 1) 텍스트 데이터 시각화
 - KoNLPy (형태소 분석기) 사용

```python
from konlpy.tag import Hannanum
engine=Hannanum()

text='나는 대학생이다. 내가 듣는 수업은 빅데이터 시각화 과목이다'

print(engine.analyze(text))
print(engine.morphs(text))
print(engine.nouns(text))
print(engine.pos(text))
```

## 워드클라우드
 - 1) Collections 라이브러리

```python
from collections import Counter
fruit_list=['사과', '딸기', '바나나', '사과', '딸기', '바나나', '바나나', '수박', '바나나','사과']
result=Counter(fruit_list)
print(result)
print(result.most_common(2))
```

- 2) 워드클라우드 그리기

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
tags=result.most_common(4)

wordcloud=WordCloud(font_path='c:\Windows\Fonts\malgun.ttf',
                   background_color='white',
                   width=1200, height=800).generate_from_frequencies(dict(tags))
fig=plt.figure(figsize=(12, 12))
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```

- 3) 2022 윤석열 대통령 취임사

```python
from konlpy.tag import Hannanum
from collections import Counter
import matplotlib.pyplot as plt
from wordcloud import WordCloud

text=open('2022_president_message.txt', 'r').read()
engine=Hannanum()
nouns_list=engine.nouns(text)
nouns=[n for n in nouns_list if len(n)>1]
count=Counter(nouns)
tags=count.most_common(100)

wordcloud=WordCloud(font_path='c:\Windows\Fonts\malgun.ttf', background_color='white', width=1200, height=800).generate_from_frequencies(dict(tags))
fig=plt.figure(figsize=(12,12))
plt.axis('off')
plt.imshow(wordcloud)
plt.show()
```
